Importing Library 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import *
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import sklearn.tree
from sklearn.tree import DecisionTreeClassifier 
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report ,confusion_matrix ,ConfusionMatrixDisplay
from sklearn.metrics import confusion_matrix
```

Importing Dataset

```python
data = pd.read_csv('/content/drive/MyDrive/Iris(1).csv')
```

Data Visualisation

```python
plt.figure(figsize=(7,5))
sns.heatmap(data.corr(), annot=True)
plt.show()
```

Building Model

```python
labelencoder = LabelEncoder()
X = data.iloc[:,5].values
data.iloc[:,5] = labelencoder.fit_transform(X)
X = X.reshape(-1,1) 
```

```python
data
```

```python
x1 = data.drop(['Species'], axis='columns', inplace=False)
y1 = data['Species']

X_train, X_test, y_train, y_test = train_test_split(x1, y1, test_size=0.33, random_state = 0)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
Decision_tree = DecisionTreeClassifier()
Decision_tree.fit(X_train,y_train)
```

```python
prediction = Decision_tree.predict(X_test)
prediction 
```

Decision tree visualization 

```python
from sklearn import tree
fn=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
cn=['setosa', 'versicolor', 'virginica']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (15,10), dpi=300)
tree.plot_tree(Decision_tree,
               feature_names = fn, 
               class_names=cn,
               filled = True);
fig.savefig('imagename.png')
```

Accuracy And Score

```python
score=accuracy_score(y_test,prediction)
print("Accuracy:",score)
```

```python
target_names = ['Iris-virginica','Iris-setosa','Iris-versicolor']
```

```python
matrix = confusion_matrix(y_test,prediction, labels=[1,0])
print('Confusion matrix : \n',matrix)

tp, fn, fp, tn = confusion_matrix(y_test,prediction,labels=[1,0]).reshape(-1)
print('Outcome values : \n', tp, fn, fp, tn)

matrix = classification_report(y_test,prediction,labels=[1,0])
print('Classification report : \n',matrix)
```

Confusion Matrix

```python
cm = confusion_matrix(y_test, prediction.reshape(-1,1), normalize='true')
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=target_names)
fig, ax = plt.subplots(figsize=(8,6))
disp.plot(ax=ax,cmap='Blues')
```
